Comparing DCGAN, WGAN AND ACGAN on Performance Using FID Score

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DCGAN = np.load('Results/FID_score/DC_FID.npy')
WGANGP = np.load('Results/FID_score/WGAN_FID.npy')
ACGAN = np.load('Results/FID_score/ACGAN_FID.npy')
```

```python
plt.figure(figsize=(10, 5))
plt.title("FID Scores Comparisons: DCGAN vs WGAN vs ACGAN")
plt.plot( DCGAN, label="DCGAN", color='red')
plt.plot( WGANGP, label="WGAN", color='green')
plt.plot( ACGAN, label="ACGAN", color='blue')
plt.xlabel("Epochs")
plt.ylabel("FID Score")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('Results/FID_summary_final.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
plt.show()
```
